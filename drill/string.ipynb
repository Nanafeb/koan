## 文字列からデータを作成

```
id,first_name,last_name,email
1,Ada,McGrale,amcgrale0@livejournal.com
2,Astra,Bullcock,abullcock1@sun.com
3,Domenic,Lippett,dlippett2@sun.com
4,Barbie,Thieme,bthieme3@tiny.cc
5,Ximenez,Isabell,xisabell4@unicef.org
```
このような、一行目はヘッダー、あとの行はデータの文字列を、

```python
[{'email': 'amcgrale0@livejournal.com',
  'first_name': 'Ada',
  'id': '1',
  'last_name': 'McGrale'},
 {'email': 'abullcock1@sun.com',
  'first_name': 'Astra',
  'id': '2',
  'last_name': 'Bullcock'},
 {'email': 'dlippett2@sun.com',
  'first_name': 'Domenic',
  'id': '3',
  'last_name': 'Lippett'},
 {'email': 'bthieme3@tiny.cc',
  'first_name': 'Barbie',
  'id': '4',
  'last_name': 'Thieme'},
 {'email': 'xisabell4@unicef.org',
  'first_name': 'Ximenez',
  'id': '5',
  'last_name': 'Isabell'}]
```
このように、ヘッダーをキーに持つ辞書データのリストに変換してください。

```python
s="""id,first_name,last_name,email
1,Ada,McGrale,amcgrale0@livejournal.com
2,Astra,Bullcock,abullcock1@sun.com
3,Domenic,Lippett,dlippett2@sun.com
4,Barbie,Thieme,bthieme3@tiny.cc
5,Ximenez,Isabell,xisabell4@unicef.org"""

string_to_data(s) 

# [{'email': 'amcgrale0@livejournal.com',
#   'first_name': 'Ada',
#   'id': '1',
#   'last_name': 'McGrale'},
#  {'email': 'abullcock1@sun.com',
#   'first_name': 'Astra',
#   'id': '2',
#   'last_name': 'Bullcock'},
#  {'email': 'dlippett2@sun.com',
#   'first_name': 'Domenic',
#   'id': '3',
#   'last_name': 'Lippett'},
#  {'email': 'bthieme3@tiny.cc',
#   'first_name': 'Barbie',
#   'id': '4',
#   'last_name': 'Thieme'},
#  {'email': 'xisabell4@unicef.org',
#   'first_name': 'Ximenez',
#   'id': '5',
#   'last_name': 'Isabell'}]


s="""first_name,tel,car owned
Dion,41-958-9263,Jeep
Joe,30-348-6500,Nissan
Sibelle,42-177-8432,Dodge"""

string_to_data(s) 

# [{'car': 'Jeep', 'first_name': 'Dion', 'tel': '41-958-9263'},
#  {'car': 'Nissan', 'first_name': 'Joe', 'tel': '30-348-6500'},
#  {'car': 'Dodge', 'first_name': 'Sibelle', 'tel': '42-177-8432'}]
```

上記をデータをヘッダーをKey、各行をValueにもつ辞書のリストに変換して下さい。ただし、
+ company nameは
    + 大文字に変換
+ revenueは、
    + 通貨記号は削除
    + B: billion = 10億 で数値変換
    + M: millon = 100万 で数値変換
    + n/a: は numpy nan に変換
+ stock priceは
    + 小数点1位まで（四捨五入）
