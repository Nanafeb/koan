**メタ文字**

メタ文字|説明|指定例|合致する
---|---|---|---
.|任意の一文字|a.c|abc, acc, aac
^|行の先頭|^abc|abcdef
$|行の末尾|abc$|defabc
*|0回以上の繰り返し|ab*|a, ab, abb, abbb
+|1回以上の繰り返し|ab+|ab, abb, abbb
?|0回または1回|ab?|a, ab
{m}|m回の繰り返し|a{3}|aaa
{m,n}|m〜n回の繰り返し|a{2, 4}|aa, aaa, aaaa
[★]|★のどれか1文字|[a-c]|a, b, c
★|★|★のどれか|a|b|a, b


**特殊シーケンス**

特殊シーケンス|説明|同じ意味の正規表現
---|---|---
\d|任意の数字|[0-9]
\D|任意の数字以外|[^0-9]
\s|任意の空白文字|[\t\n\r\f\v]
\S|任意の空白文字以外|[^\t\n\r\f\v]
\w|任意の英数字|[a-xA-Z0-9_]
\W|任意の英数字以外|[\a-xA-Z0-9_]
\A|文字列の先頭|^
\Z|文字列の末尾| $


```
# 先頭が ネコ であるか match メソッドを使って確認する
# 返り値が <_sre.SRE_Match object; span=(0, 2), match='ネコ'> であればOK

s = "ネコ 犬 ぞう mouse ネコ"

```

```
# sのぞうを象に書き換えてください。
# 'ネコ 犬 象 mouse ネコ'　が出力されれば正解です。
s = "ネコ 犬 ぞう mouse ネコ"

```

```
# s に mouse が存在するかを、 re.search をつかって確認してください
# <_sre.SRE_Match object; span=(8, 13), match='mouse'>　が出力されれば正解です

s = "ネコ 犬 ぞう mouse ネコ"

```

```
# s から ネコを全部取り出してください。
# ['ネコ', 'ネコ'] が出力されれば正解です。
s = "ネコ 犬 ぞう mouse ネコ"
```
